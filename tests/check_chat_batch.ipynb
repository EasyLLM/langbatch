```python
from langbatch import OpenAIChatCompletionBatch

batch = OpenAIChatCompletionBatch.create(
    [
        [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "Hello, how are you?"}
        ],
        [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "What can you do?"}
        ]
    ],
    model="gpt-4o-mini"
    )

id = batch.id
print(id)
```

```python
batch = OpenAIChatCompletionBatch("1253a691-fc77-4c81-bff7-33469001ba49.jsonl")
```

```python
batch.save()
loaded_batch = OpenAIChatCompletionBatch.load(id)

print(loaded_batch.id)
print(loaded_batch._file)
print(loaded_batch.platform_batch_id)
```

```python
batch.start()
```

```python
batch.get_status()
```

```python
if batch.get_status() == "completed":
    results_file = batch.get_results_file()
    print(results_file)
```

```python
success, failed = batch.get_results()

for result in success:
    print(result["custom_id"], result["choices"][0]["message"]["content"])

for result in failed:
    print(result["custom_id"], result["error"])
```
