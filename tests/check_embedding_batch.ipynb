```python
from langbatch import OpenAIEmbeddingBatch

batch = OpenAIEmbeddingBatch.create(
    [
        "I did not have a good day yesterday",
        "I am feeling good today",
        "I am feeling okay today",
    ],
    model="text-embedding-ada-002"
    )
```

```python
batch = OpenAIEmbeddingBatch("27a76bdb-3fa1-4b4a-b992-813162b0f5e6.jsonl")
```

```python
batch.start()
```

```python
batch.status()
```

```python
if batch.status() == "completed":
    results_file = batch.get_results_file()
    print(results_file)
```

```python
success, failed = batch.get_results()

for result in success:
    print(result["custom_id"], result["embedding"])

for result in failed:
    print(result["custom_id"], result["error"])
```
